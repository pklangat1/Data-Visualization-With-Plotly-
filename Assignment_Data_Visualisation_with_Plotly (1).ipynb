```python
#Data Cleaning and stardardazation
# We now rename our columns.
# We can use this method if we have many column names.
# We will use the str.strip(), str.lower(), str.replace() functions 
# to ensure that our column names are in lowercase format that we easily 
# reference while performing further analysis.
# ---
# str.strip() - This fuction is used to remove leading and trailing characters.
# str.lower() - This function is used to convert all characters to lowercase
# str.replace() - This fuction is used to replace text with some other text.
# ---
#
mobile_df.columns = mobile_df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# Then preview our resulting dataframe
# ---
mobile_df.head()
```

```python
# Prepare: Creating a summary table  
summary_total = mobile_df.groupby(['country_name'], as_index=False).agg({'1996': np.sum})
summary_total
```

```python
# Render
fig = px.bar(summary_total,
    x = "country_name", 
    y = "1996",
    labels = {
        'x' : "country_name", 
        'y' : "demo"
    },  
  )

# Tweak and Label
fig.update_layout(
    title = {
        'text': "Mobile data for year of 1996", 
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'}, 
    xaxis_title = "Year of !996 Mobile data",
    yaxis_title = "Countries",
    hoverlabel = dict(
        bgcolor = "white",
        font_size = 16,
        font_family = "Rockwell"
    )
  )

# Display
fig.show()

```

```python
# Example 2
# ---
# Question: Create a horizontal bar chart to compare tips for each countries.
# ---
# Explanation
# - `orientation='h'` sets the bar chart orientation to horizontal.
# - `color` automatically applies color to the bars.

# Prepare: Creating a summary table
summary_series = mobile_df.groupby(['series_name'], as_index=False).agg({'2016': np.sum})
summary_series
```

```python
# Render
fig = px.bar(summary_series,
    x = "series_name", 
    y = "2016",
    labels = {
        'x' : "Series Name", 
        'y' : "Year Of 2016"
    },
    orientation = 'h',
    color = "series_name"
  )

# Tweak and Label
fig.update_layout(
    title = {
        'text': "Total Tips for each Series name", 
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'}, 
    xaxis_title = "Total Series Name",
    yaxis_title = "Gender",
    hoverlabel = dict(
        bgcolor = "white",
        font_size = 16,
        font_family = "Rockwell"
    )
  )

# Display
fig.show()
```

```python
# Question
# ---
# Plot a histogram to display the distribution of the Series Name in the given dataset.
```

```python
# Render
fig = px.line(
    mobile_df, 
    x = "country_name",
    y = "series_name",
    height = 500, 
    width = 800,
    color_discrete_sequence = px.colors.qualitative.Pastel
  )

# Tweak and Label
fig.update_layout(
    title = {
        'text': 'Series name for each Since 1995', 
        'x': 0.5,
        'font_size' : 15,
        'xanchor': 'center',
        'yanchor': 'top'},
    xaxis_title_text = 'Countries',
    yaxis_title_text = 'Series Name'
  ) 

# Display
fig.show()
```
